```python
# Load env variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Create an API Client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
# Make a request
message = client.messages.create(
    model=model,
    max_tokens = 1000,
    messages=[
        {"role": "user", 
         "content": "What is the capital of France?"}
    ]
)

message
```
